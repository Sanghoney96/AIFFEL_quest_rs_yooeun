```python
import os
import cv2
import matplotlib.pyplot as plt
import numpy as np
import dlib
```

```python
my_image_path = 'images/2_suzy_frontal.jpg'
img_bgr = cv2.imread(my_image_path) # OpenCV  
img_show = img_bgr.copy() #    
plt.imshow(img_bgr) #      .
plt.show() #  .
```

```python
# plt.imshow  RGB    .
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb) #       (        .)
plt.show() #   
```

```python
detector_hog = dlib.get_frontal_face_detector() #    
```

```python
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
dlib_rects = detector_hog(img_rgb, 1)   # (image, num of image pyramid)
```

```python
#     
#     
print(dlib_rects)

for dlib_rect in dlib_rects: #    
    l = dlib_rect.left() # 
    t = dlib_rect.top() # 
    r = dlib_rect.right() # 
    b = dlib_rect.bottom() # 

    cv2.rectangle(img_show, (l,t), (r,b), (0,255,0), 2, lineType=cv2.LINE_AA) #       

img_show_rgb =  cv2.cvtColor(img_show, cv2.COLOR_BGR2RGB)
plt.imshow(img_show_rgb)
plt.show()
```

```python
model_path = 'models/shape_predictor_68_face_landmarks.dat'
    #  landmark   model_path  
landmark_predictor = dlib.shape_predictor(model_path)
    # dlib  shape_predictor    
    # landmark_predictor RGB dlib.rectangle   dlib.full_object_detection 
    # dlib.rectangle:   
    # dlib.full_object_detection:    ,     
```

```python
list_landmarks = []
    #    list 

#     face landmark 
# face landmark  
for dlib_rect in dlib_rects:
    points = landmark_predictor(img_rgb, dlib_rect)
        #  landmark  points  
    list_points = list(map(lambda p: (p.x, p.y), points.parts()))
        #  landmark  (x,y)   list_points  
    list_landmarks.append(list_points)
        # list_landmarks   

print(len(list_landmarks[0]))
    #  n  list_landmarks n  
    #   68    list
    # list_landmarks  1 list_landmarks[1]  IndexError 
```

```python
# list_landmarks  1    
for landmark in list_landmarks:
    for point in landmark:
        cv2.circle(img_show, point, 2, (0, 255, 255), -1)
            # cv2.circle: OpenCV   
            # img_show    point
            #  2 (0, 255, 255)  (-1)  
            # (       )

img_show_rgb = cv2.cvtColor(img_show, cv2.COLOR_BGR2RGB)
    # RGB  
plt.imshow(img_show_rgb)
    #  
plt.show()
    #  
```

```python
for dlib_rect, landmark in zip(dlib_rects, list_landmarks): #      68      
    print (landmark[30]) #  index 30 
    x = landmark[30][0] #    x
    y = landmark[30][1] #- dlib_rect.height()//2 #    y -      //2  (437, 182-(186+1//2))
    w = h = dlib_rect.width() #       (531-345+1)  max(x) - min(x) +1(    1       1 )
    print (f'(x,y) : ({x},{y})')
    print (f'(w,h) : ({w},{h})')
```

```python
sticker_path = 'images/cat-whiskers.png' #   
img_sticker = cv2.imread(sticker_path) #    // cv2.imread( )  image  
img_sticker = cv2.resize(img_sticker, (w,h)) #     w,h      (187) // cv2.resize(image , ( ,  ))
print (img_sticker.shape) #      
```

```python
# x,y,w,h      
# x    x = 438
# y    y = 251
# w       = 322
# h       = 322
refined_x = x - w // 2
refined_y = y - h // 2
#       x, y  .   top-left  .
# , refined_x, refined_y   
print (f'(x,y) : ({refined_x},{refined_y})') #   :   
#       ,  -         
# opencv ndarray , ndarray        .
```

```python
#        
#   y,x . (y,x,rgb channel)
#   -y   crop , top  x y             .
#    (    ) .
if refined_x < 0:
    img_sticker = img_sticker[:, -refined_x:]
    refined_x = 0
#        y  
if refined_y < 0:
    img_sticker = img_sticker[-refined_y:, :] # refined_y -98, img_sticker[98: , :] . (187, 187, 3) (89, 187, 3)  (187  98 )
    refined_y = 0

print (f'(x,y) : ({refined_x},{refined_y})')
```

```python
#sticker_area     crop  .
#  (344,0)  (344+187, 0+89)   .
#   y,x . (y,x,rgb channel)
# img_show[refined_y:refined_y+img_sticker.shape[0], refined_x:refined_x+img_sticker.shape[1]]
# img_show[0:0+89, 344:344+187]
# img_show[0:89, 344:531]
# , x 344~531 / y 0~89 
sticker_area = img_show[refined_y:refined_y+img_sticker.shape[0], refined_x:refined_x+img_sticker.shape[1]]
#   0    ,    .
#   0 (  )  
# sticker_area        .
# np.where       .
#   img_sticker 0 (      ) sticker_area(       ) ,
#   img_sticker( ) .
img_show[refined_y:refined_y+img_sticker.shape[0], refined_x:refined_x+img_sticker.shape[1]] = \
    np.where(np.all(img_sticker == 255, axis=-1, keepdims=True), sticker_area, img_sticker).astype(np.uint8)
```

```python
#     .
#  (7-3) (7-4)   img_show      .)
plt.imshow(cv2.cvtColor(img_show, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
#  
sticker_area = img_bgr[refined_y:refined_y +img_sticker.shape[0], refined_x:refined_x+img_sticker.shape[1]]
# img_bgr 7-2 rgb   .
img_bgr[refined_y:refined_y +img_sticker.shape[0], refined_x:refined_x+img_sticker.shape[1]] = \
    np.where(np.all(img_sticker == 255, axis=-1, keepdims=True), sticker_area, img_sticker).astype(np.uint8)
plt.imshow(cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)) # rgb          .
plt.savefig('_', dpi=300)
plt.show()
```
